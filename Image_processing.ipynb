# Grayscaling

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
```

```python
image = np.zeros((256,256,3),np.uint8)
image[:,:,2]=255*np.ones((256,256),np.uint8)
image[:,:,1]=255*np.ones((256,256),np.uint8)
image[:,:,0]=255*np.ones((256,256),np.uint8)
```

```python
cv2_imshow(image)
```

```python
img
```

```python
img = cv2.resize(img,(400,400))
img = cv2.resize(img,(img.shape[1]//2,img.shape[0]//2))
cv2_imshow(img)
```

```python
img.shape
```

# Flipping an image


```python
flip_img = cv2.flip(img,-1)
cv2_imshow(flip_img)
```

```python
img.shape
```

```python
crop_img = img[100:150,50:100]
cv2_imshow(crop_img)
```

# saving an image


```python
cv2.imwrite('cv2.png',crop_img)
```

# Drawing shapes

```python
image = np.zeros((512,512,3)) #canvas
```

```python
cv2_imshow(image)
```

```python
rec_img = cv2.rectangle(image,pt1=(200,200),pt2=(100,100),color=(255,255,0),thickness=1)
```

```python
cv2_imshow(rec_img)
```

```python
circle = cv2.circle(image,center=(100,400),radius=50,color=(100,0,255),thickness=-1)
```

```python
cv2_imshow(circle)
```

```python
line=cv2.line(image,pt1=(0,0),pt2=(512,512),color=(0,100,100),thickness=2)
cv2_imshow(line)
```

```python
line=cv2.putText(image,org=(200,400),fontScale=4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA,text="Hello",fontFace=cv2.FONT_ITALIC)
cv2_imshow(line)
```
