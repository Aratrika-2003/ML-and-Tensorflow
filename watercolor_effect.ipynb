Clear impurities

```python
clear = cv2.medianBlur(img,3)
clear = cv2.medianBlur(img,3)
clear = cv2.medianBlur(img,3)
clear = cv2.edgePreservingFilter(clear,sigma_s = 5)
```

```python
cv2_imshow(clear)
```

# Filtering images

```python
filter = cv2.bilateralFilter(clear,3,10,5)

for i in range(2):
  filter = cv2.bilateralFilter(filter,3,20,10)

for i in range(3):
  filter = cv2.bilateralFilter(filter,5,30,10)
```

```python
cv2_imshow(filter)
```

Tuning art

```python
gaussian_mask = cv2.GaussianBlur(filter,(7,7),2)
sharp = cv2.addWeighted(filter,1.5,gaussian_mask,-0.5,0)
```

```python
cv2_imshow(sharp)
```
